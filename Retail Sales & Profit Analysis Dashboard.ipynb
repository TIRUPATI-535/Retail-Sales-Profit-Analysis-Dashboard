```python
dataset['Profit']=dataset['Sales']- dataset['Cost']
dataset['Profit']
```

```python
dataset["Profit Margin"]=(dataset["Profit"]/ dataset["Sales"])*100
dataset["Profit Margin"]
```

```python
dataset["Total_Discount"]=dataset["Quantity"]*dataset["Unit Price"]*dataset["Discount"]
dataset["Total_Discount"]
```

```python
# Average revenue(Sales) of the each product
avg_revenue = dataset.groupby('Product')['Sales'].mean()
avg_revenue
```

```python
# Total revenue of the each product
Total_Revenue=dataset.groupby("Product")["Sales"].sum()
Total_Revenue
```
